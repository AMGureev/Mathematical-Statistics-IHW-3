### Пункт 1.

МНК-оценка параметров  $a$  и  $b$  
Создадим вектор $c$, где $c = (X^TX)^{-1}X^TY$, $c = (a, b)$

```python
Y = np.array(array)
X = np.array([np.array([1, i+1]) for i in range(200)])
c = np.linalg.inv((X.transpose() @ X)) @ X.transpose() @ Y
print(f"Полученные оценки параметров:\na = {c[0]}\nb = {c[1]}")
```

### Пункт 2.

Оценка дисперсии $σ_ε^2$  
Формула дисперсии  $σ_ε^2 = \frac{RSS}{n-p-1}$  
$RSS = \sum^n_{i=1}({y_i-\hat y_i})^2$

```python
n = 200 # кол-во наблюдений
p = 2 # кол-во параметров
Y_pred = X @ c
res = Y - Y_pred
RSS = np.sum(res ** 2)
sigma = RSS / (n - p - 1)
print(f"Полученная оценка дисперсии:\nRSS = {RSS}\nsigma = {sigma}")
```

### Пункт 3.

Оценка коэффициента детерминации  
Формула вычисления коэф. детерминации: $R^2 = 1 - \frac{RSS}{TSS}$  
$TSS = \sum^n_{i=1}({y_i-\hat y})^2$

```python
TSS = np.sum((Y - np.mean(Y)) ** 2)
deter = 1 - (RSS/TSS)
print(f"Коэф. детерминации = {deter}")
if deter > 0.8:
  print("Коэф. детерминации близок к 1, значит модель точная")
else:
  print("Коэф. детерминации слишком мал, значит модель неточная")
```

### Пункт 4.

Проверьте гипотезу $H_0:b=0$

Оценка коэффициентов линейной регрессии с использованием метода наименьших квадратов:  
$\hat{\beta} = (X^{\top}X)^{-1}X^{\top}y$

Предсказанные значения: $\hat{y} = X \hat{\beta}$

Остатки: $e = y - \hat{y}$

Оценка дисперсии ошибки:  
$\hat{\sigma}^2 = \frac{\sum_{i=1}^{n} e_{i}^{2}}{n - p}$  
где $ n $ — количество наблюдений, $ p $ — количество параметров модели (включая свободный член).

Стандартные ошибки для коэффициентов:  
$SE(\hat{\beta}_{j}) = \sqrt{\frac{\hat{\sigma}^2}{(X^{\top}X)^{-1}_{jj}}}$

t-статистика для коэффициента $\beta_{j}$:  
$T = \frac{\hat{\beta_{j}}}{SE(\hat{\beta_{j}})}$

Критическое значение t для уровня значимости $\alpha$ и степеней свободы $n - p - 1$:  
$t_{critical} = t_{\alpha/2, n-p-1}$
где $t_{\alpha/2, n-p-1}$ — квантиль распределения Стьюдента.

Гипотезы:  
1) $H_{0}: \beta = 0$  
2) $H_{1}: \beta \neq 0$

```python
X_trans_X_inv = np.linalg.inv(X.T @ X)
errors = np.sqrt(np.diag(sigma * X_trans_X_inv))

t_stats = c / errors

t_critical_1 = sci.stats.t.ppf(0.95 / 2, df = n - p)
t_critical_2 = sci.stats.t.ppf(1 - 0.95 / 2, df = n - p)
print(f"Коэф. b: {c[1]}")
print(f"t-статистика: {t_stats[1]}")
print(f"Критические значения t: {t_critical_1}; {t_critical_2}")
```

### Пункт 5.

Построить точечную и интервальную оценку (уровня надёжности 0.95)  

Модель с оценёнными параметрами:  
$\hat{Y} = \hat{b}_{0} + \hat{b}_{1} t_{?}$

Для вычисления интервальной оценки $ Y $ в точке $ t_{?} $, используем следующие формулы:

1. Факт распределения статистики:  
$\frac{Y - \mathbb{E}[\hat{Y}]}{\sqrt{\mathbb{D}[\hat{Y}]}} \sim t(n - p - 1)$

2. Интервальная оценка:  
$\mathbb{P}\left( \hat{Y} - t_{n-p-1}^{0.025} \sqrt{\mathbb{D}[\hat{Y}] + \mathbb{E}[\hat{Y}]} < Y < \hat{Y} + t_{n-p-1}^{0.025} \sqrt{\mathbb{D}[\hat{Y}] + \mathbb{E}[\hat{Y}]} \right) = 0.95$

Для нашей модели:  
$\mathbb{E}[\hat{Y}] = \hat{b}_{0} + \hat{b}_{1} t_{?}$   
$\mathbb{D}[\hat{Y}] = \sigma_{e}^{2} \left( \frac{1}{n} + \frac{(t_{?} - \bar{t})^{2}}{\sum_{i=1}^{n} (t_{i} - \bar{t})^{2}} \right)$

Здесь $\hat{b}_{0}$ и $\hat{b}_{1}$ — оценённые коэффициенты регрессии, $\sigma_{e}^{2}$ — оценка дисперсии ошибки, $n$ — количество наблюдений, $p$ — количество параметров модели, $\bar{t}$ — среднее значение $t_{i}$ в выборке.

```python
t_100 = 100
t_200 = 200
res_100 = c[0] + c[1] * t_100
res_200 = c[0] + c[1] * t_200
print(f"Точечная оценка для t=100: {res_100}")
print(f"Точечная оценка для t=200: {res_200}")
```

```python
t_sum = sum(range(1, n + 1)) / n
sum_squared_diff = sum((i - t_sum) ** 2 for i in range(1, n + 1))
D_100 = sigma * (1 + 1/n + (t_100 - t_sum) ** 2 / sum_squared_diff)
D_200 = sigma * (1 + 1/n + (t_200 - t_sum) ** 2 / sum_squared_diff)
L_100 = t_critical_1 * res_100
L_200 = t_critical_1 * res_200
R_100 = t_critical_1 * (D_100 ** 0.5) + res_100
R_200 = t_critical_1 * (D_200 ** 0.5) + res_200
print(f'Интервал для t=100: ({L_100}; {R_100})')
print(f'Интервал для t=200: ({L_200}; {R_200})')
```

### Пункт 7.

С помощью критерия Голдфелда-Квандта проверьте (на уровне значимости 0.05) гипотезу о гомоскедастичности модели.  
  
  
Воспользуемся критерием Голдфелда-Квандта для проверки гипотезы о гомоскедастичности модели.

Гипотезы:
- $H_0: \sigma^2_1 = \dots = \sigma^2_n$
- $H_a: \sigma^2_i = \sigma^2_0 X_{ij}$

Сначала упорядочим данные по $t$ в порядке убывания и исключим середину (при $d = \frac{n}{4}$). После этого разделим данные на две части и построим регрессии для каждой части, оценивая:
$\hat\epsilon_1^{[I]}, \dots, \hat\epsilon_{\frac{n-d}{2}}^{[I]}$
$\hat\epsilon_1^{[II]}, \dots, \hat\epsilon_{\frac{n-d}{2}}^{[II]}$

Статистика теста:
$T = \frac{\sum{(\hat\epsilon_i^{[I]})^2}}{\sum{(\hat\epsilon_i^{[II]})^2}} \Bigg|_{H_0} \sim \text{F}\left(\frac{n-d}{2}-p-1, \frac{n-d}{2}-p-1\right)$

Доверительная область:
$(0; f^{1-0.05}_{\frac{n-d}{2}-p-1, \frac{n-d}{2}-p-1})$

Применим этот критерий к нашим данным, чтобы проверить гипотезу о равенстве дисперсий:

```python
data = list(zip(range(1, n+1), Y))[::-1]
d = n // 4
subset0 = data[:n//2 - d//2]
subset1 = data[n//2 + d//2:]

def create_matrices(subset):
    Y = np.array([value for _, value in subset])
    X = np.array([[1, index] for index, _ in subset])
    return X, Y

X0, Y0 = create_matrices(subset0)
X1, Y1 = create_matrices(subset1)

B0 = np.linalg.inv(X0.T @ X0) @ X0.T @ Y0
B1 = np.linalg.inv(X1.T @ X1) @ X1.T @ Y1
b00, b01 = B0
b10, b11 = B1

residuals0 = [value - b00 - b01 * index for index, value in subset0]
residuals1 = [value - b10 - b11 * index for index, value in subset1]
RSS0 = sum(residual ** 2 for residual in residuals0)
RSS1 = sum(residual ** 2 for residual in residuals1)
T0 = RSS0 / RSS1

critical_value = sci.stats.f.ppf(1 - 0.05, n//2 - d//2 - p - 1, n//2 - d//2 - p - 1)
lower_bound = 0

print(f'T0 = {T0}\nДоверительный интервал = ({lower_bound}; {critical_value})')
```

### Пункт 8.

Рассмотрите модель $Y_i = a + bt_i + ct_i^2 + ε_i,  i = 1, ... ,200.$ На уровне
значимости 0.05 решите вопрос о том, следует ли включать в модель
регрессор $t^2$.  

Построим обе регрессии и проверим гипотезу
$H_0: t_i^2 = 0$  
$H_a: t_i^2 \neq 0$

Статистика
$$T = \frac{\text{RSS}^{[II]} - \text{RSS}^{[I]}}{\frac{\text{RSS}^{[I]}}{n-p-1}} \Big|_{H_0} \sim \text{F}(1, n-p-1)$$

Доверительная область
$$(0; f^{1-0.05}_{1;n-p-1})$$

```python
Xe = np.array([[1] * n, range(1, n+1), [i**2 for i in range(1, n+1)]]).transpose()
Be = np.linalg.inv(Xe.T @ Xe) @ Xe.T @ Y
Ye = Xe @ Be
RSSe = sum((Y[i] - Ye[i]) ** 2 for i in range(n))
T0 = (RSSe - RSS) / (RSS / (n - p - 1))
critical_value = sci.stats.f.ppf(1 - 0.05, 1, n - p - 1)
lower_bound = 0
print(f'T0 = {T0}\nДоверительный интервал = ({lower_bound}; {critical_value})')
```

Гипотеза отвергается, следовательно, новый регрессор улучшает модель (*отрицательное значение статистики указывает на то, что новая модель более точно предсказывает значения.*)
